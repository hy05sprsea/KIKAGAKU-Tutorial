```
from PIL import Image
```

```
img = Image.open('sample.png')
```

```
type(img)
```

```
img
```

```
img = np.array(img)
```

```
img.shape
```

```
img.dtype
```

```
img_gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
```

```
plt.imshow(img_gray)
```

```
plt.imshow(img_gray, cmap='gray')
```

```
# エッジ検出のフィルタの定義（横方向）
kernel = np.array([
    [-1, 0, 1],
    [-1, 0, 1],
    [-1, 0, 1]
])
```

```
# 畳み込み演算
img_conv = cv2.filter2D(img_gray, -1, kernel)
```

```
# 結果を可視化
plt.imshow(img_conv, cmap='gray');
```

```
# エッジ検出のフィルタの定義（縦方向）
kernel = np.array([
    [-1, -1, -1],
    [0, 0, 0],
    [1, 1, 1]
])
```

```
# 畳み込み演算
img_conv = cv2.filter2D(img_gray, -1, kernel)
```

```
     
# 結果を可視化
plt.imshow(img_conv, cmap='gray');
```

```
# エッジ検出のフィルタの定義（縦方向）
kernel = np.array([
    [1, 1, 1],
    [0, 0, 0],
    [-1, -1, -1]
])

# 畳み込み演算
img_conv = cv2.filter2D(img_gray, -1, kernel)

# 結果を可視化
plt.imshow(img_conv, cmap='gray');
```
